```julia
using DelimitedFiles

# ファイルを読み込む
data = readdlm("chushutu.dat", ',')

# １列目の値を1/1000倍する
data[:, 1] .= data[:, 1] .*1000

# 結果を保存するファイル名
output_file = "modified_chushutu.dat"

# 変更を保存する
writedlm(output_file, data, ',')
```

```julia
# data.datのデータを読み込む
data = readdlm("UGa2.dat", ' ')

# 区切り文字を,に変更したdata2.datファイルにデータを書き込む
writedlm("dat2.dat", data, ',')
```

```julia
using Plots

p = plot(dpi=1000)

data_UGa2 = readdlm("UGa2.dat", '\t', Float64)

# データを x_UGa2 と y_UGa2 に分割
x_UGa2 = data_UGa2[:, 1]
y_UGa2 = data_UGa2[:, 2]

# yのピークでの値が1になるように倍率を計算
scale_factor_UGa2 = 1 / maximum(y_UGa2)

# yを倍率でスケーリング
y_scaled_UGa2 = y_UGa2 * scale_factor_UGa2

# プロットに追加
plot!(p, x_UGa2, y_UGa2, xlabel="Binding energy (eV)", ylabel="Intensity (arb.unit)", label="UGa2.dat (Scaled)", marker=:auto, markersize=0)

display(p)
```
